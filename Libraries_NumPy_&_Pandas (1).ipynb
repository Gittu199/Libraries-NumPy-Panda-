```python
# =============================================================
# Libraries: NumPy & Pandas Assignment
# =============================================================
# Name: Libraries NumPy & Pandas Assignment_35
# Dataset: Retail Sales Dataset
# =============================================================

# ---------------------------------------------------------------
# SETUP: Install and Import Required Libraries
# ---------------------------------------------------------------
# !pip install pandas numpy   # Uncomment if running on Google Colab

import pandas as pd
import numpy as np
```

```python
# ---------------------------------------------------------------
# Q1. Load the given retail sales dataset into a Pandas DataFrame
#     and display the first 5 rows along with the dataset
#     information to understand its structure.
# ---------------------------------------------------------------

# Load the dataset
# If running on Google Colab, upload the CSV file first using:
# from google.colab import files
# uploaded = files.upload()
# df = pd.read_csv('retail_dataset_.csv')

df = pd.read_csv('/content/retail_dataset .csv')
```

```python
print("=" * 60)
print("Q1: First 5 Rows of the Dataset")
print("=" * 60)
print(df.head())

print("\nDataset Information:")
print("-" * 40)
df.info()

print("\nDataset Shape (Rows, Columns):", df.shape)
print("\nColumn Names:", df.columns.tolist())
```

```python
df.head(5)
```

```python
# ---------------------------------------------------------------
# Q2. Inspect the dataset and identify missing values in each
#     column. Display the total number of missing values present
#     in the dataset.
# ---------------------------------------------------------------

print("\n" + "=" * 60)
print("Q2: Missing Values in Each Column")
print("=" * 60)
missing_per_column = df.isnull().sum()
print(missing_per_column)

total_missing = df.isnull().sum().sum()
print(f"\nTotal Missing Values in the Dataset: {total_missing}")
```

```python
# ---------------------------------------------------------------
# Q3. Replace missing values in the Quantity and Price columns
#     with their respective mean values using Pandas.
# ---------------------------------------------------------------

print("\n" + "=" * 60)
print("Q3: Replacing Missing Values with Mean")
print("=" * 60)

# Mean of Quantity and Price (ignoring NaN)
quantity_mean = df['Quantity'].mean()
price_mean = df['Price'].mean()

print(f"Mean of Quantity: {quantity_mean:.2f}")
print(f"Mean of Price:    {price_mean:.2f}")

# Replace missing values with mean
df['Quantity'] = df['Quantity'].fillna(quantity_mean)
df['Price'] = df['Price'].fillna(price_mean)

print("\nMissing values after replacement:")
print(df[['Quantity', 'Price']].isnull().sum())
print("Missing values in Quantity and Price have been successfully replaced.")
```

```python
# ---------------------------------------------------------------
# Q4. Remove all rows where the Product Category or Region
#     column contains missing values.
# ---------------------------------------------------------------

print("\n" + "=" * 60)
print("Q4: Removing Rows with Missing Product Category or Region")
print("=" * 60)

print(f"Shape before dropping: {df.shape}")
df = df.dropna(subset=['Product Category', 'Region'])
print(f"Shape after dropping:  {df.shape}")
print("Rows with missing Product Category or Region have been removed.")
```

```python
# ---------------------------------------------------------------
# Q5. Create a new column called Revenue using NumPy, where:
#     Revenue = Quantity × Price
#     Display the updated dataset with the new column.
# ---------------------------------------------------------------

print("\n" + "=" * 60)
print("Q5: Creating Revenue Column using NumPy")
print("=" * 60)

df['Revenue'] = np.multiply(df['Quantity'].values, df['Price'].values)

print("Updated Dataset (first 10 rows with Revenue column):")
print(df[['Transaction ID', 'Product Category', 'Quantity', 'Price', 'Revenue']].head(10))
print(f"\nRevenue column successfully added. Total rows: {len(df)}")
```

```python
# ---------------------------------------------------------------
# Q6. Using NumPy, calculate the total revenue generated from
#     all transactions in the dataset.
# ---------------------------------------------------------------

print("\n" + "=" * 60)
print("Q6: Total Revenue using NumPy")
print("=" * 60)

total_revenue = np.nansum(df['Revenue'].values)
print(f"Total Revenue Generated from All Transactions: ₹{total_revenue:,.2f}")
```

```python
# ---------------------------------------------------------------
# Q7. Group the dataset by Product Category and calculate the
#     total revenue generated by each category.
# ---------------------------------------------------------------

print("\n" + "=" * 60)
print("Q7: Total Revenue by Product Category")
print("=" * 60)

category_revenue = df.groupby('Product Category')['Revenue'].sum().reset_index()
category_revenue.columns = ['Product Category', 'Total Revenue']
category_revenue = category_revenue.sort_values('Total Revenue', ascending=False)
print(category_revenue.to_string(index=False))
```

```python
# ---------------------------------------------------------------
# Q8. Based on the revenue calculated for each product category,
#     identify the top 3 categories with the highest revenue and
#     the bottom 3 categories with the lowest revenue.
# ---------------------------------------------------------------

print("\n" + "=" * 60)
print("Q8: Top 3 & Bottom 3 Product Categories by Revenue")
print("=" * 60)

cat_rev_series = df.groupby('Product Category')['Revenue'].sum().sort_values(ascending=False)

top_3 = cat_rev_series.head(3)
bottom_3 = cat_rev_series.tail(3)

print("\nTop 3 Categories with Highest Revenue:")
print("-" * 40)
for i, (cat, rev) in enumerate(top_3.items(), 1):
    print(f"  {i}. {cat}: ₹{rev:,.2f}")

print("\nBottom 3 Categories with Lowest Revenue:")
print("-" * 40)
for i, (cat, rev) in enumerate(bottom_3.items(), 1):
    print(f"  {i}. {cat}: ₹{rev:,.2f}")
```

```python
# ---------------------------------------------------------------
# Q9. Group the dataset by Region and calculate the total revenue
#     generated in each region. Identify the region with the
#     highest revenue and the region with the lowest revenue.
# ---------------------------------------------------------------

print("\n" + "=" * 60)
print("Q9: Total Revenue by Region")
print("=" * 60)

region_revenue = df.groupby('Region')['Revenue'].sum().sort_values(ascending=False)

print("Revenue by Region:")
print("-" * 40)
for region, rev in region_revenue.items():
    print(f"  {region}: ₹{rev:,.2f}")

highest_region = region_revenue.idxmax()
lowest_region = region_revenue.idxmin()

print(f"\nRegion with Highest Revenue: {highest_region}  →  ₹{region_revenue[highest_region]:,.2f}")
print(f"Region with Lowest Revenue:  {lowest_region}  →  ₹{region_revenue[lowest_region]:,.2f}")
```

```python
# ---------------------------------------------------------------
# Q10. Extract the month from the Date column using Pandas and
#      group the dataset by month to calculate the total monthly
#      revenue. Also, calculate the mean, median, and standard
#      deviation of the Revenue column using NumPy.
# ---------------------------------------------------------------

print("\n" + "=" * 60)
print("Q10: Monthly Revenue & Revenue Statistics")
print("=" * 60)

# Extract month from Date column
df['Date'] = pd.to_datetime(df['Date'], dayfirst=True)
df['Month'] = df['Date'].dt.month

# Group by Month
month_names = {1:'January', 2:'February', 3:'March', 4:'April',
               5:'May', 6:'June', 7:'July', 8:'August',
               9:'September', 10:'October', 11:'November', 12:'December'}

monthly_revenue = df.groupby('Month')['Revenue'].sum()

print("Total Revenue by Month:")
print("-" * 40)
for month, rev in monthly_revenue.items():
    print(f"  {month_names[month]:<12}: ₹{rev:,.2f}")

# NumPy statistics
rev_values = df['Revenue'].values
mean_rev    = np.nanmean(rev_values)
median_rev  = np.nanmedian(rev_values)
std_rev     = np.nanstd(rev_values)

print("\nRevenue Statistics (using NumPy):")
print("-" * 40)
print(f"  Mean Revenue:               ₹{mean_rev:,.2f}")
print(f"  Median Revenue:             ₹{median_rev:,.2f}")
print(f"  Standard Deviation:         ₹{std_rev:,.2f}")

print("\n" + "=" * 60)
print("Assignment Completed Successfully!")
print("=" * 60)
```
